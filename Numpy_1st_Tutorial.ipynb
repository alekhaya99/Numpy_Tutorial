# Multidimensional array object

```
#Generates a random matrix of 2*3
data=np.random.rand(2,3)
data
```

```
#Multiply an integer a float with the array elements
data*10
```

```
#Adds two matrices together
data+data
```

```
#Convert the input data (list, tuple, array, or other sequence type to an nd array) either infering a d type or explicitly specifying a d type; copies the input data by default
a=[1,2,3,4,5,6]
np.array(a)
```

```
#Convert input to an nd array, but do not copy the input is already in the nd-array
a=[1,2,3,4,5,6]
np.asarray(a)
```

```
#like the builtin range, the following returns an nd array in that specified range
np.arange(10)
```

```
#The following produces an array of all ones with the given shape and d-type
a = np.ones([2,4], dtype = float) 
print("\nMatrix a : \n", a) 
  
```

```
#The following takes another array and produces a ones array of the same shape and d-type
np.ones_like(a)
```

```
#like ones and one like, the following produces zero
b = np.zeros([2,4], dtype = float) 
print("\nMatrix b : \n", b) 
np.zeros_like(b)
```

```
#Creates new arrays by allocating new memory, but do not populates with value like ones or zeros
c=np.empty([2,3],dtype=int)
print(c)
```

```
np.empty_like(c)
```

```
#Produces an array of the given shape and dtype with all value set to indicated value
d= np.full([2, 2], [32,67], dtype = int) 
print("\nMatrix d : \n", d) 
  
e = np.full([3, 3], 10.1) 
print("\nMatrix e : \n", e) 
```

```
np.full_like(d,[9,11])
```

```
'''R : Number of rows
C : [optional] Number of columns; By default M = N
k : [int, optional, 0 by default]
          Diagonal we require; k>0 means diagonal above main diagonal or vice versa.
dtype : [optional, float(by Default)] Data type of returned array.  '''
g = np.eye(2, dtype = float) 
print("Matrix g : \n", g) 
  
# matrix with R=4 C=5 and 1 on diagnol 
# below main diagnol 
h = np.eye(4, 5, k = -1) 
print("\nMatrix h : \n", h)
```

```
#Creates an identity matrix
np.identity(3)
```
